```python
import json
import random
import time


with open('test_documents_v2.json', 'r') as file:
    data = json.load(file)

offset = 3
last_index = 0
chunk_number = 1
file_path = "C:/Users/talha/Documents/YouGov/llm-enhanced-pipeline/data/docs_bucket"


while last_index < len(data):
    doc_content = []
    for item in data[last_index:last_index + offset]:
        doc_content.append(item)

    print(f"Content: {doc_content}\n")


    output_filename = f"/test_doc_v{chunk_number}.json"
    with open(file_path+output_filename, 'w', encoding='utf-8') as outfile:
        json.dump(doc_content, outfile, indent=2)
    print(f"Written to {output_filename}\n")

    last_index += offset
    chunk_number += 1
    time.sleep(random.randint(1,5))
```

```python
import re

def fetching_numbers_from_text(text):
    match = re.search(r"(\d+)", text)
    if match:
        return match.group(1)
    else:
        return None
```

```python
doc = data[0]

def processed_doc(doc):
    processed_doc = {}

    processed_doc["document_id"] = doc["document_id"]
    processed_doc["raw_text"] = doc["text"]
    processed_doc["invoice_number"] =  fetching_numbers_from_text(doc["text"])
    processed_doc["shop"] = doc["shop"].upper()
    processed_doc["total"] = doc["total"].split()[0]
    processed_doc["currency"] = doc["total"].split()[-1]
    processed_doc["timestamp"] = doc["timestamp"]
    processed_doc["items"] = doc["purchase"]

    return processed_doc
```

```python
import numpy as np


def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


def text_cosine_similarity(text1, text2):
    # Step 1: convert text to vectors
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([text1, text2])

    # Step 2: compute cosine similarity
    similarity = cosine_similarity(vectors[0], vectors[1])

    # Step 3: return score
    return similarity[0][0]


# Example usage
text1 = "9355 BookPlanet"
text2 = "9355 BookPlanet".upper()

score = text_cosine_similarity(text1, text2)

print("Similarity:", score)
```
